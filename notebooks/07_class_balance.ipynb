```python
import pandas as pd
import os
from imblearn.over_sampling import SMOTE

# CARGA DE DATOS
# USAR '../' para salir de la carpeta notebooks e ir a data
X_train = pd.read_csv("..\datos\processed\X_train_fe.csv")
y_train = pd.read_csv("..\datos/processed\y_train.csv")

# ENCODING 0 y 1

X_train_numeric = pd.get_dummies(X_train)

# PROCESO DE BALANCEO (SMOTE)
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train_numeric, y_train.values.ravel())

# EXPORT DATAFRAME FINAL
df_balanced = pd.DataFrame(X_train_res, columns=X_train_numeric.columns)
df_balanced['Revenue'] = y_train_res

# GUARDAR EN LA RUTA: data\processed
output_dir = "../data/processed"
output_file = "X_train_balanced_final.csv"
full_path = os.path.join(output_dir, output_file)

# CREAR CARPETA SI NO EXISTIERA
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df_balanced.to_csv(full_path, index=False)

print("--- REPORTE FINAL ---")
print(f"✅ Archivo guardado en: {full_path}")
print(f"📊 Balance Final: {pd.Series(y_train_res).value_counts().to_dict()}")
```
