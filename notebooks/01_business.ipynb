# GLOSARIO DE TÉRMINOS

**Comportamiento de Navegación (Métricas de Sesión)**

Administrative / Informational / ProductRelated: 
Número de páginas de cada tipo visitadas por el usuario en esa sesión.

Administrative_Duration / Informational_Duration / ProductRelated_Duration: 
Tiempo total (en segundos) que el usuario permaneció en cada categoría de página.

BounceRates: 
Porcentaje de visitantes que entran al sitio y se van ("rebotan") sin realizar ninguna otra acción.

ExitRates: 
Porcentaje de visitas a la página que fueron las últimas en la sesión.

PageValues: 
Valor promedio de la página visitada antes de completar una transacción (es el predictor con mayor impacto).

**Información del Visitante y Temporalidad**

SpecialDay: 
Indica qué tan cerca está la fecha de la sesión a un día especial (ej. San Valentín, Día de la Madre). Rango de 0 a 1.

Month / Weekend: 
Mes de la sesión y si esta ocurrió en fin de semana (variable binaria).

OperatingSystems / Browser / Region / TrafficType: 
Datos técnicos y geográficos sobre el origen y la tecnología del visitante.

VisitorType: 
Clasifica si el cliente es "Nuevo", "Recurrente" u "Otro".

**Variable Objetivo (Target)**
Revenue: Variable binaria que indica si la sesión culminó en una transacción final (compra).

# KPIS DE ÉXITO DEL MODELO
Debido a que solo el 15.5% de las sesiones terminan en compra (tenemos clases desbalanceadas), el Accuracy no servirá de mucho. 
Se sugiere usar F1-Score o AUC-ROC.

Impacto esperado:
+ tasa de conversión
+ ingreso por visitante 
- costo de adquisición

# VARIABLE OBJETIVO
Al querer convertir intenciones de compra en compras reales, el target es la columna Revenue.

# PROBLEMA A RESOLVER
El objetivo principal es desarrollar un modelo de clasificación capaz de predecir si una sesión de navegación de un usuario terminará en una transacción (`Revenue = True`). 

Esto permitirá al negocio tomar decisiones proactivas para mejorar la tasa de conversión.

# CONTEXTO DE LA INDUSTRIA
En el e-commerce, entender por qué un cliente abandona el sitio sin comprar es crítico. Al predecir la intención de compra, el equipo de marketing puede:
* Ofrecer cupones de descuento personalizados en tiempo real.
* Mejorar la asistencia mediante chatbots si se detecta indecisión.
Esta ejecución en tiempo real se traduce en maximizar inversiones y mejorar la experiencia de usuario

# TÍTULO Y DESCRIPCIÓN DEL PROYECTO
Título: Análisis de Intención de Compra en E-commerce
Dataset: Online Shoppers Purchasing Intention
Descripción:Dataset de UCI Machine Learning Repository con sesiones web de una tienda online durante 1 ano. Cada fila es una sesión unica de un usuario distinto. 
Solo el 15.5 % de las sesiones terminaron en compra. Incluye metricas de Google Analytics y datos del visitante.
